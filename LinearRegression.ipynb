```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
true_w = 2
true_b = 3
x = np.linspace(0,20,50)
noise = np.random.randn(50) * 2
y = true_w * x + true_b + noise

def compute_cost(x,y,w,b):
  total_cost = 0
  m = x.shape[0]
  for i in range(m):
    predicted = w * x[i] + b
    cost =  (y[i]- predicted) ** 2
    total_cost += cost
  return total_cost/(2*m)


```

```python
def compute_gradient(x,y,w,b):
  m= x.shape[0]
  dj_dw = 0
  dj_db = 0
  for i in range(m):
    error = (w * x[i] + b) - y[i]
    dj_db = dj_db + error
    dj_dw = dj_dw + error * x[i]
  return dj_dw/m , dj_db/m

```

```python
def gradient_descent(x,y,w_init,b_init,alpha,iterations):
  w = w_init
  b = b_init
  cost_history = []
  for i in range(iterations):
    dj_dw,dj_db = compute_gradient(x,y,w,b)
    w = w - alpha * dj_dw
    b = b - alpha * dj_db
    cost = compute_cost(x,y,w,b)
    cost_history.append(cost)
    if i%100 == 0:
      print(f"Iterations: {i}, Cost: {cost}, w: {w}, b: {b}")
  return w,b,cost_history


w_init = 0
b_init = 0
alpha = 0.01
iterations = 1000

w_final,b_final,cost_history = gradient_descent(x,y,w_init,b_init,alpha,iterations)
```

```python
plt.plot(cost_history)
plt.xlabel("Iterations")
plt.ylabel("cost")
plt.title("cost over iterations")
plt.show()
```

```python
plt.scatter(x, y, color="red", marker="x", label="Data points")
plt.plot(x, w_final * x + b_final, color="blue", label="Fitted line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression Fit")
plt.show()

# Output final parameters
print(f"Final parameters: w = {w_final}, b = {b_final}")
```
